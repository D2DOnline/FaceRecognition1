```python
def create_lookup(words):
  lookup={}
  for i in range(len(words)):
    word= words[i].decode('utf-8')
    lookup[word]=i  
  return lookup
```

```python
words_lookup = create_lookup(words)
```

```python
def generate_id_representation(review_words):
  word_ids = []
  current_len = 0
  for word in review_words:
    if current_len >= MAX_SEQUENCE_LENGTH:
      break
    if word in words_lookup:
      word_ids.append(words_lookup[ word] )
    else:
      word_ids.append(0)  
  current_len = current_len + 1 
  if current_len < MAX_SEQUENCE_LENGTH:
    for i in range(current_len, MAX_SEQUENCE_LENGTH):
      word_ids.append(0)   
  return word_ids 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
all_reviews['text_ids'] = all_reviews['words'].map(generate_id_representation)
text_ids = all_reviews['text_ids' ].values.tolist()
x_input = np.array(text_ids)
y_labels = np.array(all_reviews['sentiment'].tolist())
np.save('datasets/ids_matrix.npy',text_ids)
np.save('datasets/ids_matrix.npy' , text_ids)
x_train,x_test, y_train, y_test = train_test_split(x_input, y_labels, test_size=0.2)
saved_embeddings = np.load('datasets/wordVectors.npy')
saved_embeddings.shape
NUM_EPOCHS = 15
BATCH_SIZE = 25
EMBEDDING_SIZE = 50
MAX_LABEL = 2
tf.reset_default_graph()
x =tf.placeholder(tf.int32, [None, MAX_SEQUENCE_LENGTH])
y = tf.placeholder(tf.int32, [None])
embeddings = tf.nn.embedding_lookup(saved_embeddings, x)
```

```python
gruCell = tf.nn.rnn_cell.GRUCell(EMBEDDING_SIZE)
gruCell = tf.nn.rnn_cell.DropoutWrapper(cell=gruCell, output_keep_prob=0.66)
_, encoding = tf.nn.dynamic_rnn(gruCell, embeddings, dtype=tf.float32)
logits = tf.layers.dense(encoding, MAX_LABEL, activation=None)
cross_entropy = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits, labels = y)
loss = tf.reduce_mean(cross_entropy)
prediction = tf.equal(tf.cast(tf.argmax(logits, 1), tf.int32), y)
accuracy = tf.reduce_mean(tf.cast(prediction, tf.float32))
optimizer = tf.train.AdamOptimizer(0.001)
train_step = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
with tf.Session() as session:
  init.run()

  for epoch in range(NUM_EPOCHS):

    x_dataset = tf.data.Dataset.from_tensor_slices(x_test)
    y_dataset = tf.data.Dataset.from_tensor_slices(y_train)
  
    batched_dataset = tf.data.Dataset.zip((x_dataset, y_dataset)) .batch(BATCH_SIZE)
    iterator = batched_dataset.make_one_shot_iterator()
    next_element = iterator.get_next()
    num_batches=int(len(x_train) // BATCH_SIZE)+1
    for i in range(num_batches):
      x_train_batch,y_train_batch=session.run(next_element)

      train_dict={x:x_train_batch, y:y_train_batch}
      session.run(train_step,feed_dict=train_dict)
      train_loss, train_acc=session.run([loss,accuracy],feed_dict=train_dict)
    test_dict={x:x_test, y:y_test}
    test_loss, test_acc=session.run([loss,accuracy], feed_dict=test_dict)

    print ('Epoch: {}, Train Loss: {: .2}, Train Acc: { .5}' .format (epoch +1, train_loss, train_acc))
    print ('Epoch: {}, Test Loss: {: .2}, Test Acc: { .5}' .format (epoch +1, test_loss, test_acc))    
```
